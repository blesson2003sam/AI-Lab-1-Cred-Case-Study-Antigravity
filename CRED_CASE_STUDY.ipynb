A list is a built-in Python data structure used to store multiple values in a single variable.

**Characteristics:**


*  Lists are ordered
*  Lists are mutable
*  Lists allow duplicate values






A list is defined using square brackets [].

```python
numbers = [10, 20, 30]
```

```python
customer_id = ["cid-0011", "cid-0012", "cid-0013", "cid-0014", "cid-0015"]
```

```python
print(customer_id)
```

```python
credit_score = [650, 720, 580, 690, 710]
age = [25, 30, 28, 35, 40]
debt_status = ["No Debt", "No Debt", "Debt", "No Debt", "No Debt"]
active_credit_cards = [1, 3, 2, 4, 1]
prec_on_time = [0.80, 0.75, 0.30, 0.90, 0.60]
```

```python
print(customer_id)
print(credit_score)
print(age)
print(debt_status)
print(active_credit_cards)
print(prec_on_time)
```

Python lists start indexing from 0.
The first element has index 0, second has index 1, and so on.

```python
index = customer_id.index("cid-0012")

print(customer_id[index])
print(credit_score[index])
print(age[index])
print(debt_status[index])
print(active_credit_cards[index])
```

```python
index = customer_id.index("cid-0011")
debt_status[index] = "Debt"

print(debt_status)
```

```python
if prec_on_time[1] > 0.7:
    credit_score[1] = credit_score[1] * 1.02

print(credit_score)
```

```python
for i in range(len(prec_on_time)):
    if prec_on_time[i] > 0.75:
        credit_score[i] = credit_score[i] * 1.02

print(credit_score)
```

```python
for i in range(len(prec_on_time)):
    if prec_on_time[i] < 0.40:
        credit_score[i] = credit_score[i] * 0.98

print(credit_score)
```

```python
count = 0
for score in credit_score:
    if score > 700:
        print(score)
        count += 1

print("Total customers with score > 700:", count)
```

```python
count = 0
for cards in active_credit_cards:
    if cards > 2:
        count += 1

print("Customers with more than 2 cards:", count)
```

```python
debt = 0
no_debt = 0

for status in debt_status:
    if status == "Debt":
        debt += 1
    else:
        no_debt += 1

print("Debt:", debt)
print("No Debt:", no_debt)
```

```python
for i in range(len(customer_id)):
    if customer_id[i] == "cid-0012":
        print(customer_id[i], credit_score[i], age[i], debt_status[i], active_credit_cards[i])
```

```python
def get_customer_details(cust_id):
    for i in range(len(customer_id)):
        if customer_id[i] == cust_id:
            print(customer_id[i], credit_score[i], age[i], debt_status[i], active_credit_cards[i])
```

```python
get_customer_details("cid-0011")
```

```python
get_customer_details("cid-0012")
```

```python
customer_data = {
    "cust_id": customer_id,
    "credit_score": credit_score,
    "age": age,
    "debt_status": debt_status,
    "active_credit_cards": active_credit_cards
}

print(customer_data)
```
